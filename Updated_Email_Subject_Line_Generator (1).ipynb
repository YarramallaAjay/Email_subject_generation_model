```python
import os
import pandas as pd

# File paths for train and test data directories
train_dir_path = '/content/AESLC/enron_subject_line/train'
test_dir_path = '/content/AESLC/enron_subject_line/test'

def process_subject_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        subject_text = file.read().strip()
        return {'File': os.path.basename(file_path), 'Subject': subject_text}

def create_csv(directory_path, output_csv):
    data = []
    for filename in os.listdir(directory_path):
        if filename.endswith('.subject'):
            file_path = os.path.join(directory_path, filename)
            data.append(process_subject_file(file_path))

    df = pd.DataFrame(data)
    df.to_csv(output_csv, index=False)

# Create CSV files from directories
output_csv_train = "/content/train_output.csv"
output_csv_test = "/content/test_output.csv"

create_csv(train_dir_path, output_csv_train)
create_csv(test_dir_path, output_csv_test)

# Load the CSV file into a DataFrame
df_train = pd.read_csv(output_csv_train)
df_test = pd.read_csv(output_csv_test)

# Preprocessing function
def preprocess_df(df):
    df['Text'] = df['Subject'].apply(lambda x: x.split('@subject')[0].strip() if '@subject' in x else '')
    df['Sub'] = df['Subject'].apply(lambda x: x.split('@subject')[1].strip() if '@subject' in x else '')
    df.drop(columns=['Subject', 'File'], inplace=True)
    return df

# Apply preprocessing to train and test DataFrames
df_train = preprocess_df(df_train)
df_test = preprocess_df(df_test)

# Check the first few rows
```

```python
import pandas as pd
from datasets import Dataset
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load T5 model and tokenizer
checkpoint = "t5-base"
tokenizer = T5Tokenizer.from_pretrained(checkpoint)
model = T5ForConditionalGeneration.from_pretrained(checkpoint)

# # Load preprocessed CSV files
# train_df = pd.read_csv('/content/train_output.csv')
# test_df = pd.read_csv('/content/test_output.csv')

# Convert DataFrame to Dataset
train_dataset = Dataset.from_pandas(df_train)
test_dataset = Dataset.from_pandas(df_test)

# Verify columns after conversion
print("Train dataset columns:", train_dataset.column_names)
print("Test dataset columns:", test_dataset.column_names)

# Tokenization function
def preprocess_function(dataSets):
    # Ensure the expected columns are present
    if 'Text' not in dataSets or 'Sub' not in dataSets:
        raise ValueError("Expected columns 'Text' and 'Sub' not found in examples")

    # Tokenize input texts and target labels
    inputs = ["summarize: " + doc for doc in dataSets["Text"]]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding='max_length')

    # Tokenize the target (subject line)
    labels = tokenizer(dataSets["Sub"], max_length=30, truncation=True, padding='max_length')

    # Ensure labels are the same length as inputs
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Apply preprocessing to train and test datasets
# Remove columns after ensuring their existence
train_dataset = train_dataset.map(preprocess_function, batched=True)
test_dataset = test_dataset.map(preprocess_function, batched=True)

# Verify dataset after processing
print("Processed train dataset columns:", train_dataset.column_names)
print("Processed test dataset columns:", test_dataset.column_names)
```

```python
def generate_response(model, tokenizer, prompt):
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to("cuda")

    # Create the attention mask and pad token id and move to cuda
    attention_mask = torch.ones_like(input_ids).to("cuda")
    pad_token_id = tokenizer.eos_token_id

    # Move the model to cuda
    model.to("cuda")

    output = model.generate(
        input_ids,
        max_new_tokens=100,
        num_return_sequences=1,
        attention_mask=attention_mask,
        pad_token_id=pad_token_id
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example email body
email_body = "The 2002 Enron Net Works budget will for the first time combine allocations to business units for four key services:  Enron Online, IT Infrastructure, Risk Management Operations and IT Maintenance and Development. We are providing a level of detail for expense and capital spend that is greater than that provided in past years, and that will allow business units to make informed choices regarding the services that they receive.We are meeting with individual business unit Offices of the Chair to review in detail the four service areas delivered through Enron Net Works.Some of these meetings have been completed and more will occur between now and October 12.Please keep the following points in mind as we work with you to finalize budget:  Enron Networks today is very different from this tim"

# Tokenize and generate subject
# Move inputs to cuda
inputs = tokenizer.encode("summarize: " + email_body, return_tensors="pt", max_length=512, truncation=True).to("cuda")

# Move model to cuda
model.to("cuda")

outputs = model.generate(inputs, max_length=30, num_beams=4, early_stopping=True)

# Decode the generated subject
subject = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Subject:", subject)
```

```python
def extract_subject(subject_text):
    # Convert subject_text to string to handle potential float values
    subject_text = str(subject_text)
    start = subject_text.find('@subject') + len('@subject')
    end = subject_text.find('@subject', start)
    if end != -1:
        return subject_text[start:end].strip()
    else:
        return subject_text[start:].strip()
```

```python
def get_subject(subject_text):
  extracted_text = extract_subject(subject_text).split('<EMAIL_ENDED>', 1)[0].strip()
  return extracted_text
```

```python
from evaluate import load
from collections import defaultdict
from tqdm import tqdm

def metric(data):
    sub,ann0,ann1,ann2,gen_subject  = ["Subject","Annotation 0","Annotation 1","Annotation 2","Generated Subject"]
    rouge = load("rouge")
    meteor = load("meteor")
    score = defaultdict(int)

    for _,x in tqdm(data.iterrows(),total=len(data)):

        rouge_s = rouge.compute(predictions=[x[gen_subject]],references=[x[sub]])
        rouge_h = rouge.compute(predictions=[x[gen_subject]] ,references=[[x[ann0],x[ann1],x[ann2]]])

        score["rouge_1_sub"] += rouge_s['rouge1']
        score["rouge_2_sub"] += rouge_s['rouge2']
        score["rouge_L_sub"] += rouge_s['rougeL']

        score["rouge_1_human"] += rouge_h['rouge1']
        score["rouge_2_human"] += rouge_h['rouge2']
        score["rouge_L_human"] += rouge_h['rougeL']


        meteor_s = meteor.compute(predictions=[x[gen_subject]],references=[x[sub]])
        meteor_h = meteor.compute(predictions=[x[gen_subject]] ,references=[[x[ann0],x[ann1],x[ann2]]])

        score["meteor_sub"] += meteor_s["meteor"]
        score["meteor_human"] += meteor_h["meteor"]

    print("The evalution scores are as follows:")

    for name,val in score.items():
        print(f"{name}: {val/len(data)}")
```
